## Data preprocessing and cleaning

```python
#Changing host verification to binary
listings.host_identity_verified[listings.host_identity_verified == 't'] = 1
listings.host_identity_verified[listings.host_identity_verified == 'f'] = 0

#Changing instant bookable to binary
listings.instant_bookable[listings.instant_bookable == 't'] = 1
listings.instant_bookable[listings.instant_bookable == 'f'] = 0
```

```python
#We do not want listings that have not yet been reviewed, so they are removed
listings = listings[listings.number_of_reviews != 0]


#Dropping listings which has no name,hostname,description of bathrooms and no total review score
listings = listings[listings.name.notna()]
listings = listings[listings.host_name.notna()]
listings = listings[listings.bathrooms_text.notna()]
listings = listings[listings.review_scores_rating.notna()]

#Now checking number of null values
listings.isnull().sum()
```
